```python
#standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import ticker, cm
import seaborn as sns
from scipy import signal,ndimage,linalg

from glob import glob
import os
import gc
import wfdb
```

# Data import 

```python
filepaths=glob ("combined-measurement-of-ecg-breathing-and-seismocardiograms-1.0.0/files/*.dat")
print (filepaths)

keylist=[]
for i in range (1,21):
    if i>9:
        keylist.append("0"+str(i))
    else:
        
        keylist.append("00"+str(i))
print (keylist)
```

# Data parser loop

```python
for key in keylist:
    print("LOADING SUBJECT {0} MEASUREMENTS!".format(key))

    measurements= [string for string in filepaths if key in string]
    
    for m in measurements:
        print("LOADING:")
        rid=m.split(".dat")[0][-4:]
        print (rid)
        record = wfdb.rdrecord(rid, pn_dir='cebsdb')
        data = pd.DataFrame(record.p_signal, columns=[signal for signal in record.sig_name])
        data.columns=["ECG1","ECG2","RESPIRATION","Acc_Z"]
        print (data.head())
        print (data.shape)
        print("")
        print ("saving formatted csv: ",rid)
        data.to_csv("Formatted_CEBSDB_Data/{0}".format(rid),index=False)
```
