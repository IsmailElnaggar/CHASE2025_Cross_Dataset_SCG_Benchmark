# Data parser loop

```python

for key in keylist:
    st=time.time()
    print("LOADING SUBJECT {0} MEASUREMENTS!".format(key))

    record = wfdb.rdrecord(key, pn_dir='scg-rhc-wearable-database/processed_data/')
    data = pd.DataFrame(record.p_signal, columns=[signal for signal in record.sig_name])
    print (data.head())
    print (data.shape)
    print("")
    print ("saving formatted csv: ",key)
    data.to_csv("Formatted_RHC_Data/{0}".format(key),index=False)
    et=time.time()
    print ("DOWNLOAD COMPLETE!")
    print ("TIME: {0}".format(et-st))
```
