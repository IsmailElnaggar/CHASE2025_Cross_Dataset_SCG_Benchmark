```python
for mypath in filepaths:
    if "CP" in mypath:
        print ("loading:",mypath)
        head, tail = os.path.split(mypath)
        data=pd.read_csv("{0}".format(mypath),skiprows=[0,2],delimiter=",")
        display(data.head())
        if "Unnamed: 13" in data.columns:
        
            data.drop("Unnamed: 13",axis=1,inplace=True)
        data.columns=["Timestamp","Acc_X","Acc_Y","Acc_Z","ECG_EMG_Status1_CAL","ECG_EMG_Status2_CAL",
                     "ECG_LA_RA","ECG_LL_LA","ECG_LL_RA","ECG_Vx_RL","Gyro_X","Gyro_Y","Gyro_Z"]
        print (data.head())
        print (data.shape)
        print("")
        print ("saving formatted csv: ",mypath)
        data.to_csv("Formatted_VHD_Data/{0}".format(tail),index=False)
```
